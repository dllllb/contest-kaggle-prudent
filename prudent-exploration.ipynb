```python
from pandas import read_csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style('ticks')

pd.set_option('display.max_rows', 150)
```

```python
df = read_csv('train.csv.gz', index_col='Id', dtype={'Product_Info_2': np.str_})
```

```python
df.Response.value_counts().plot(kind='pie', figsize=(8, 8))
```

```python
from sklearn.manifold import TSNE
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction import DictVectorizer
from sklearn.preprocessing import Imputer
from sklearn.decomposition import TruncatedSVD

pl1 = Pipeline([
    ('dv', DictVectorizer(sparse=False)),
    ('fna', Imputer(strategy='median')),
    ('svd', TruncatedSVD(n_components=50, random_state=0)),
    ('tsne', TSNE(n_components=2, init='pca', random_state=0)),
])

x1, x2 = pl1.fit_transform(df.drop('Response', axis=1).to_dict(orient='records')[:1000]).T

sns.set_context(rc={"figure.figsize": (10, 8)})
plt.scatter(x1, x2, c=df.Response[:1000], alpha=0.3)
plt.show()
```

```python
mk_sum_by_resp = df.groupby('Response')[[col for col in df.columns if col.startswith('Medical_Keyword')]].sum()
resp_counts = df.Response.value_counts()
class_feature_counts = mk_sum_by_resp.div(resp_counts, axis=0).T

from pandas.tools.plotting import scatter_matrix
scatter_matrix(class_feature_counts, alpha=0.5, figsize=(16, 16), diagonal='kde')
None
```
